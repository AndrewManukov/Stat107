#  The Madden Curse

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
aBrown_df = pd.read_csv("aBrown.csv")
tBrady_df = pd.read_csv("tBrady.csv")
rGronk_df = pd.read_csv("rGronk.csv")
```

2019 Madden Cover Athlete: Antonio Brown

```python
#aBrown_df.plot(x = "Year", y = "G")
aBrown_figure = plt.figure(1)
plt.plot(aBrown_df["Year"], aBrown["GS"])
#plt.plot([2010, 2019],[aBrown_df["GS"].mean(), aBrown_df["GS"].mean()], 'k-', lw=.2)
plt.title("Antonio Brown Games Started Per Year")
plt.xlabel("Year")
plt.ylabel("Games (Max 16)")
plt.annotate('Madden 19 Cover', xy=(2018, 15), xytext=(2015, 10),
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
aBrown_df[aBrown_df["Year"] >= 2019]
```

2018 Madden Cover Athlete: Tom Brady

```python
tBrady_figure = plt.figure(2)
plt.plot(tBrady_df["Year"], tBrady_df["Y/G"])
plt.title("Tom Brady Yards Per Game")
plt.xticks(np.arange(min(tBrady_df["Year"]), max(tBrady_df["Year"])+1, 3.0))
plt.xlabel("Year")
plt.ylabel("Yards Per Game")
plt.annotate('Madden 18 Cover', xy=(2017, 272), xytext=(2014.33, 200),
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
tBrady_df
```

```python
tBrady_df[tBrady_df["Year"] >= 2018]
```

2017 Madden Cover Athlete: Rob Gronkowski

```python
rGronk_figure = plt.figure(3)
plt.plot(rGronk_df["Year"], rGronk_df["Y/G"])
plt.title("Rob Gronkowski Yards Per Game")
plt.xticks(np.arange(min(rGronk_df["Year"]), max(rGronk_df["Year"])+1, 1.0))
plt.xlabel("Year")
plt.ylabel("Yards Per Game")
plt.annotate('Madden 17 Cover', xy=(2016, 68), xytext=(2014.33, 53),
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```
