# The Madden Curse

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import ttest_rel
```

```python
aBrown_df = pd.read_csv("aBrown.csv")
tBrady_df = pd.read_csv("tBrady.csv")
rGronk_df = pd.read_csv("rGronk.csv")
vYoung_df = pd.read_csv("vYoung.csv")

aBrown_df["Y/G"] = aBrown_df["Yds"] / aBrown_df["G"]
tBrady_df["Y/G"] = tBrady_df["Yds"] / tBrady_df["G"]
rGronk_df["Y/G"] = rGronk_df["Yds"] / rGronk_df["G"]
```

# So...is this a curse or maybe something else?
Part of these curses can't be analyized because they were the product of Injury and Outside Reasons <br>

However, the part that can be analyized is in season performance.

Let's take a look at Tom Brady, Rob Gronkowski, and one other and compare there preformances before and after the Madden Cover.

One way to quantify the performance of Quarterbacks such as Tom Brady is to use QBR (Total Quarterback Rating)


```python
tBrady_df
```

```python
tBrady_df_after_cover = tBrady_df[tBrady_df["Year"] >= 2018]
tBrady_df_after_cover
```

```python
tBrady_df_before_cover = tBrady_df[(tBrady_df["Year"] < 2018) & (tBrady_df["Year"] > 2005)]
tBrady_df_before_cover
```

```python
tBrady_before = tBrady_df_before_cover
#tBrady_n_before = len(tBrady_df_before_cover)

tBrady_after = tBrady_df_after_cover
#tBrady_n_after = len(tBrady_df_after_cover)
```

Using this data...lets run a paired T-test

```python
tBrady_one_season_before = tBrady_before[tBrady_before["Year"] >= 2016]
tBrady_one_season_after = tBrady_after[tBrady_after["Year"] <= 2019]
tBrady_one_season_after
```

(Need the sample size to be the same)

```python
ttest_rel(tBrady_one_season_before["QBR"], tBrady_one_season_after["QBR"])
```
