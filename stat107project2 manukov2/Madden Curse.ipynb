# So...is this a curse or maybe something else?

Lets do some stats! It's difficult to rate performance of players among variying positions so lets just look at the QBs <br>

Why don't we take a look at AV (Approximate Value) among all Madden Cover QB's Since 2000

Let's create 2 Data Frames. One with QBs prior to winning cover athelete and one after.



```python
qbBefore.head()
```

```python
qbAfter.head()
```

Lets take a random sample of N = 40

```python
sample_before = qbBefore.sample(n = 40)
sample_after = qbAfter.sample(n = 40)
```

Lets dive in!

```python
print("Mean Yards Before: " + str(round(sample_before["Yds"].mean(), 2)))
print("Mean Yards Per Game Before: " + str(round(sample_before["Y/G"].mean(), 2)))
print("Mean Touch Downs Before: " + str(round(sample_before["TD"].mean(), 2)))
print("Mean Interceptions Before: " + str(round(sample_before["Int"].mean(), 2)))
print("Mean AV Before: " + str(round(sample_before["AV"].mean(), 2)))
```

```python
print("Mean Yards After: " + str(sample_after["Yds"].mean()))
print("Mean Yards Per Game After: " + str(round(sample_after["Y/G"].mean(), 2)))
print("Mean Touch Downs After: " + str(round(sample_after["TD"].mean(), 2)))
print("Mean Interceptions After: " + str(round(sample_after["Int"].mean(), 2)))
print("Mean AV After: " + str(round(sample_after["AV"].mean(), 2)))
```

It's clear there is some sort of curse. Lets run a paired sample t-test on AV with alpha = .05 <br>

Null: AV before = AV after (or AV before - AV after = 0) <br>
Alternate: AV before > AV after (or AV before - AV after != 0)

```python
p_value = ttest_rel(sample_after["AV"], sample_before["AV"])[1]
print("P-value: " + str(p_value))
```

So...at alpha = .05 we reject the NULL... <br>

So the Madden Curse is real?

![yesbutno.jpg](pictures/yesbutno.jpg)

It makes sense that on average player performance is down compared to when they were selected for the Cover Athlete. Just like the MVP award, being seleted for a Cover Athlete means you've had one of the best seasons that year. It's no surprise that the years after having a "best season" performance goes down. This is a concept called Regression to the Mean. Lets see for ourselves by comparing our "after" stat to the average performance of the top 40 2018 QB's.
